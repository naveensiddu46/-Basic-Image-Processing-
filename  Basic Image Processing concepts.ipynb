# Reading Image

```python
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.image as mpimg
```

```python
img = mpimg.imread('eagle.jpg')
```

```python
img.shape
```

```python
plt.imshow(img)
plt.show()
# cv2.imshow('taj',img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
r, g, b = cv2.split(img)
```

```python
plt.figure(figsize=(15,8))
plt.subplot(131)
plt.imshow(r, cmap ='gray')
plt.title('Red Region')

plt.subplot(132)
plt.imshow(g, cmap ='gray')
plt.title('Green Region')

plt.subplot(133)
plt.imshow(b, cmap ='gray')
plt.title('Blue Region')
plt.show()

plt.imshow(img)
plt.title('Blue Green Red Region')
plt.show()
```

```python
# merging
img_new = cv2.merge((b,g,r))
```

```python
plt.figure(figsize=(15,8))
plt.subplot(131)
plt.imshow(r, cmap ='gray')
plt.title('Blue Region')

plt.subplot(132)
plt.imshow(g, cmap ='gray')
plt.title('Gree Region')

plt.subplot(133)
plt.imshow(b, cmap ='gray')
plt.title('Red Region')
plt.show()

plt.imshow(img_new)
plt.title('Blue Green Red Region')
plt.show()
```

# Converting into Grayscale

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gray.shape
```

```python
plt.imshow(gray,cmap = 'gray')
plt.show()
```

```python
cv2.imwrite('grayimag.png',gray)
```

```python
gray.min(), gray.max()
```

# Thresholding

```python
ret, thresh1 = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(gray,127,255,cv2.THRESH_BINARY_INV)
```

```python
plt.figure(figsize=(10,6))
plt.subplot(121)
plt.imshow(thresh1,cmap='gray')
plt.title('BINARY')
plt.subplot(122)
plt.imshow(thresh2,cmap='gray')
plt.title('BINARY INV')
plt.show()
```
